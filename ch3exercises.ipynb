## Exercise 2

Carefully explain the differences between the KNN classifier and KNN regression methods.

> **Solution.** We first recall the KNN classifier and KNN regression methods:
> * For the KNN classifer, given a positive $K$ and a test observation $x_0$, the KNN classifer identifies the $K$ points in the training data set that are closest to $x_0$, represented by $\mathcal{N}_0$. It then estimates the conditional probability for class $j$ as the fraction of points in $\mathcal{N}_0$ whose response values equal $j$: $$\mathrm{Pr}(Y = j | X = x_0) = \frac{1}{K} \sum_{i \in \mathcal{N}_0} I(y_i = j).$$ Finally, KNN classifies the test observation $x_0$ to the class with the largest probability.
> * For the KNN regression method, it indentifies the $K$ training observations that are closest to $x_0$, represented by $\mathcal{N}_0$, it then estimates $f(x_0)$ using the **average** of all the training responses in $\mathcal{N}_0$. I.e., $$\hat f(x_0) = \frac{1}{K} \sum_{x_i \in \mathcal{N}_0} y_i.$$
> A key difference between the KNN classifer and the KNN regression method, is that the KNN classifier assimilates the point of interest to the class with the largest probability, whereas the KNN regression method averages across a set of data. However, in a sense, the KNN classifier is producing an "average" being the expected class the given point should be a part of. For KNN regression, the optimal value of $K$ depends on the bias-variance tradeoff, whereas for the KNN classifer, we aim to choose $K$ to give a decision boundary close to the Bayes decision boundary. **Q.E.D.**

2) Answer (1) using test rather than training RSS.
3) Suppose that the true relationship between $X$ and $Y$ is not linear, but we don't know how far it is from linear. Consider the training RSS for the linear regression. 4) Answer (3) using test rather than training RSS.

> **Solution.**
> 1. Given that the true relationship between $X$ and $Y$ is linaer, we expect the RSS for cubic regression to be less than the RSS for linear regression. The linear regression is less flexible than the cubic one, and will not be able to interpolate through the training data as effectively as the cubic regression. 
> 2. However, using the **testing** data, the linear RSS will be lower than the cubic RSS. The cubic relationship, while more flexible, is more likely to overfit the data and track with noise, which affects the cubic regression's ability to demonstrate a linear trend. ## Exercise 5

Consider the fitted values that result from performing linear regression without an intercept. In this setting, the $i$th fitted value takes the form $$\hat y_i = x_i \hat\beta,$$ where $$\hat \beta = \left(\sum_{i = 1}^n x_i y_i\right)/\left(\sum_{i' = 1}^n x_{i'}^2\right).$$ Show that we can write $$\hat y_i = \sum_{i' = 1}^n a_{i'}y_{i'}.$$ What is $a_{i'}$? *Note*: We interpret this result by saying that the fitted values from linear regression are **linear combinations** of the response values.

>**Solution.** By direct computation, $$\begin{align*} \hat{y}_i &= x_i \hat \beta \\ &= x_i \frac{\sum_{i' = 1}^n x_{i'}y_{i'}}{\sum_{j = 1}^n x_{j}^2} \\ &= \sum_{i' = 1}^n \frac{x_i x_{i'}}{\sum_{j = 1}^n x_j} y_{i'}\end{align*}$$ where we see that $a = \frac{x_i x_{i'}}{\sum_{j = 1}^n x_j}$. **Q.E.D.**

## Exercise 6

Using (3.4): $$\begin{align*} \hat \beta_1 &= \frac{\sum_{i = 1}^n (x_i - \overline x)(y_i - \overline y)}{\sum_{i = 1}^n (x_i - \overline x)^2} \\ \hat \beta_0 &= \overline y - \hat \beta_1 \overline x,\end{align*}$$ argue that in the case of simple linear regression, the least squares line always passes through the point $(\overline x, \overline y)$. 

> **Solution.** In the case of simple linear regression, we have $$\hat y(x) = \hat \beta_0 + \hat \beta_1 x.$$ Substituting in $\overline x$ for $x$, and the above coefficient expressions, we obtain $$\begin{align}y(\overline x) &= (\overline y - \hat \beta_1 \overline x) + \hat \beta_1 \overline x \\ &= \overline y.\end{align}$$ As we can see, we didn't need to do much algebra, and so $y(\overline x) = \overline y$ implies the least squares line passes through the point $(\overline x, \overline y)$. **Q.E.D.**

### Part 1


```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots
```

```python
import statsmodels.api as sm
from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
auto = load_data("Auto")
```

```python
X = pd.DataFrame({'intercept' : np.ones(auto.shape[0]), 'horsepower' : auto['horsepower']})
```

```python
y = auto['mpg']
model = sm.OLS(y, X)
results = model.fit()
summarize(results)
```

- Since $P$ is low, the effect of horsepower on miles per galon is statistically significant.
- There is a very strong relationship between the predictor and the response. Observe the $P$ value is less than 0.05. 
- The relationship between the predictor and the response is negative. Generally, the higher the horsepower, the lower the vehicle's miles per gallon.
- Given a horsepower of 98, we have $$Y(98) = 39.9359 - 0.1578(98) = 24.4715 \ \mathrm{mpg}.$$

### Part 2

To plot the response and the predictor, we have,

```python
ax = auto.plot.scatter('horsepower','mpg')
def abline(ax, b, m, *args, **kwargs):
    "Add a line with slope m and intercept b to ax"
    xlim = ax.get_xlim()
    ylim = [m * xlim[0] + b, m * xlim[1] + b]
    ax.plot(xlim, ylim, *args, **kwargs)

abline(ax,
       results.params[0],
       results.params[1],
       'r--',
       linewidth=3)
```

### Part 3

We can plot the residuals $e_i = y_i - \hat y_i$ versus the predictor $x_i$ below:

```python
ax = subplots(figsize=(8,8))[1]
ax.scatter(results.fittedvalues, results.resid)
ax.set_xlabel('Fitted value')
ax.set_ylabel('Residual')
ax.axhline(0, c='k', ls='--');
```

Recall from the text that
> Ideally, the residual plot will show now discernible pattern. The presence of a pettern may indicate a problem with some aspect of the linear model.

In the plot above, we observe a general pattern with the residuals following a u-shape going from left to right. Therefore, it is unlikely that our linear model is appropriate for this data (a quadratic model is probably better).

We now plot the leverage statistics.

```python
infl = results.get_influence()
ax = subplots(figsize=(8,8))[1]
ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag)
ax.set_xlabel('Index')
ax.set_ylabel('Leverage')
np.argmax(infl.hat_matrix_diag)
```

The plot above shows a general clustering below leverage = 0.08 and a decreasing trend for data above 0.08 as the index increases. A non-linear model is likely.
